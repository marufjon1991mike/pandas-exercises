```python
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns = list('bcd'), index = ['Olma', 'Anor', 'Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape(4,3), columns = list('abc'), index = ['Olma', 'Anor', 'Qovun', 'Anjir'])
```

```python
df1
```

```python
df2
```

```python
df1 + df2 # mos qator va mos ustunlardagi qiymatlarni qo'shadi
```

```python
df1.add(df2, fill_value = 0) #add mavjud bo'lmagan qiymatlarga o'zimiz qiymat berishimiz uchun qulay. 
```

```python
df2.sub(df1, fill_value = 0)
```

```python
df1.rsub(df2, fill_value = 0) #rsub o'ngdan chapga ayirish yani df2 - df1
```
